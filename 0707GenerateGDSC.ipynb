```python
# prepare the data matrix
data_nona=data.dropna()
g_smbols = data_nona.GENE_SYMBOLS
data_nona=data_nona.drop(["GENE_title","GENE_SYMBOLS"],axis=1)
data_nona = data_nona.T
data_nona.columns = g_smbols
data_nona
```

```python
# prepare the label matrix 
label_info = label.loc[:,["COSMIC_ID","CELL_LINE_NAME","AUC","DRUG_NAME"]]
label_info
```

```python
selected_index = [d in res for d in label_info.DRUG_NAME.values]
seleced_labels = label_info[selected_index]
selected_alldrug_sens = pd.crosstab(index=seleced_labels['COSMIC_ID'], columns = seleced_labels['DRUG_NAME'],values=seleced_labels['AUC'],aggfunc='mean')
selected_alldrug_sens = selected_alldrug_sens.fillna(1)
selected_alldrug_sens
```

```python
selected_alldrug_sens.index
```

```python
transform_CID = [ "DATA."+str(i) for i in selected_alldrug_sens.index] 
```

```python
selected_alldrug_sens.index = transform_CID
selected_alldrug_sens
```

```python
select_cells = data_nona.loc[selected_alldrug_sens.index,:]
```

```python
selected_alldrug_sens = selected_alldrug_sens.drop(['DATA.908120', 'DATA.908134', 'DATA.908442', 'DATA.1789883'])
```
